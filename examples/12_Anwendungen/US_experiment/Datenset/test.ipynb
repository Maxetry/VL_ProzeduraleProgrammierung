```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url="https://raw.githubusercontent.com/TUBAF-IfI-LiaScript/VL_ProzeduraleProgrammierung/master/examples/12_Anwendungen/Datenset/distanceMeasurements_filtered.csv"

df=pd.read_csv(url, sep=',', header = 0)
df['date'] = pd.to_datetime(df['created_at'])
df['Temperature_diff']= (df["Temperature"].diff().abs() > 0.4) * 1
df.head(5)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, sharex=True)
df.plot(x="date", y="Temperature", ax= axes[0])
df.plot(x="date", y="Temperature_diff", ax= axes[1])
```
